```python
# Baca Data

import pandas as pd
import csv

data = pd.read_csv("Data_Bersih.csv")
data.columns = ['Komentar', 'Sentimen' ,'Label', 'Kategori', 'Label_Kategori']
data
```

```python
#Lowercase

import re

data["casefolding"] = [token.lower() for token in data ["Komentar"]]
data[["Komentar", "casefolding"]]
```

```python
import string

def remove_punct(text, remove_digits=True):
    pattern = r'[^a-zA-Z0-9\s\n]' if not remove_digits else r'[^a-zA-z\s\n]'
    text_nopunct =''
    text_nopunct = re.sub(pattern,'',text)
#     re.sub('['+string.punctuation+']', ' ', text)
    return text_nopunct

data['Text_noPunct'] = data['casefolding'].apply(lambda x: remove_punct(x))
data[["Komentar", "Text_noPunct"]]
```

```python
def remove_link(text, remove_digits=True):
    remove_link =''
    remove_link = re.sub(r"http\S+", "",text)
#     re.sub('['+string.punctuation+']', ' ', text)
    return remove_link

data['remove_link'] = data['Text_noPunct'].apply(lambda x: remove_link(x))
data[["Komentar", "remove_link"]]
```

```python
def open_kamus_prepro(x):
  kamus={}
  with open(x,'r') as file :
    for line in file :
      slang=line.replace("'","").split(':')
      kamus[slang[0].strip()]=slang[1].rstrip('\n').lstrip()
  return kamus
kamus_slang = open_kamus_prepro('kamus data.txt')

def slangword(text):  
    sentence_list = text.split()
    new_sentence = []
    for word in sentence_list:
      for candidate_replacement in kamus_slang:
        if candidate_replacement == word:
          word = word.replace(candidate_replacement, kamus_slang[candidate_replacement])
      new_sentence.append(word)
    return " ".join(new_sentence)
data['slang'] = data['remove_link'].apply(lambda x: slangword(x))
data[['Komentar','slang']]
```

```python
import nltk
from nltk import word_tokenize

kamus_stopword=[]
with open('stopwords.txt','r') as file :
  for line in file :
    slang=line.replace("'","").strip()
    kamus_stopword.append(slang)

def remove_stop_words(tokens):
    return [word for word in tokens if word not in kamus_stopword]

tokens = [word_tokenize(sen) for sen in data.slang]
data['stopword'] = [remove_stop_words(sen) for sen in tokens] 
data[['Komentar','stopword']]
```

```python
from nltk.stem import PorterStemmer
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()
def stemming(tokens):  
    data_stem =[]
    for i in tokens:
      kata = stemmer.stem(i)
      data_stem.append(kata)
    return data_stem

data['akhir'] = data['stopword'].apply(lambda x: stemming(x))
data[['Komentar','akhir']]
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer, TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import precision_score, accuracy_score, recall_score
from sklearn.neural_network import MLPClassifier
import pickle
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import confusion_matrix

def tfidf(data, ma = 0.6, mi = 0.0001):
    tfidf_vectorize = TfidfVectorizer(max_df = ma, min_df = mi)
    tfidf_data = tfidf_vectorize.fit_transform(data)
    pickle.dump(tfidf_vectorize, open("tfidf_model_1.pickle", "wb"))
    return tfidf_data
tfidf
```

```python
for train, test in kfold.split(x, y):
    SVM = SVC(kernel = 'linear', probability=True) #tipe kernel yang akan digunakan dalam algoritma adalah linear
    SVMClassifier = SVM.fit(x[train], y[train])
    predictions = SVMClassifier.predict(x[test])
    prediksi = SVMClassifier.predict_proba(x[test])
    vectorizer = pickle.load(open("tfidf_model_1.pickle",'rb')) #mengambil vector tfidf
    with open('svm_model_1.pickle', 'wb') as fid: #penyimpanan model
        pickle.dump(SVMClassifier, fid)

    labels = [0,1]
    prediction_labels=[]
    for p in prediksi: #for p dalam prediksi
        prediction_labels.append(labels[np.argmax(p)]) #mencari nilai tertinggi pada proba

    y_test= y[test].values
    y_tes=[]
    for i in range(0, len(y_test)): #len menghitung banyak data testing
      y_tes.append(y_test[i]) #append masukkan data testing ke array
    print(len(y_tes)) #menampilkan jumlah data testing


    y_pred=prediction_labels


    cf_sentimen = pd.DataFrame(
                data=confusion_matrix(y_tes, y_pred, labels=labels),
                columns=labels,
                index=labels
            )
    print('Confusion Matrix')
    print(cf_sentimen)

    tps_sentimen = {}
    fps_sentimen  = {}
    fns_sentimen  = {}
    tns_sentimen  = {}
    
    for label in labels:
      tps_sentimen[label] = cf_sentimen.loc[label, label]
      fps_sentimen[label] = cf_sentimen[label].sum() - tps_sentimen[label]
      fns_sentimen[label] = cf_sentimen.loc[label].sum() - tps_sentimen[label]

    for label in set(y_tes):
      tns_sentimen[label] = len(y_tes) - (tps_sentimen[label] + fps_sentimen[label] + fns_sentimen[label])

    print('True Positif Sentimen = ', tps_sentimen)
    print('False Positif Sentimen = ',fps_sentimen)
    print('False Negatif Sentimen = ',fns_sentimen)
    print('True Negatif Sentimen = ',tns_sentimen)


    accuracySentimen=sum(tps_sentimen.values())/len(y_tes)
    acc.append(accuracySentimen) #append masukkan dakurasi ke array


    tpfp_sentimen = [ai + bi for ai, bi in zip(list(tps_sentimen.values()), list(fps_sentimen.values()))]
    precision=[ai / bi if bi>0 else 0 for ai , bi in zip(list(tps_sentimen.values()), tpfp_sentimen)] #perulangan
    precisionSentimen=sum(precision)/2 #rumus presisi
    presisi.append(precisionSentimen) #append masukkan presisi ke array

    tpfn_sentimen = [ai + bi for ai, bi in zip(list(tps_sentimen.values()), list(fns_sentimen.values()))]
    recall=[ai / bi if bi>0 else 0 for ai, bi in zip(list(tps_sentimen.values()), tpfn_sentimen)]
    recallSentimen=sum(recall)/2 #rumus recall
    recal.append(recallSentimen) #append masukkan recall ke array

print('accuracy ', acc)
print('precisi ',presisi)
print('recall ',recal)
```
